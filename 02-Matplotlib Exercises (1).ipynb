## Exercise 2
** Create a figure object and put two axes on it, ax1 and ax2. Located at [0,0,1,1] and [0.2,0.5,.2,.2] respectively.**

```python
# Import the necessary libraries
import matplotlib.pyplot as plt
import numpy as np

# Create plot
fig = plt.figure(figsize=(4.2, 3))

# Create axes
ax = fig.add_axes([0, 0, 1, 1])  # Main axes
ax2 = fig.add_axes([0.2, 0.5, 0.2, 0.2])  # Inset axes

# Set tick positions
ticks = np.arange(0, 201, 50)
tick = np.arange(0, 101, 20)

# Set x-axis and y-axis limits for ax
ax.set_xlim(0, 101)
ax.set_ylim(0, 201)

# Set tick positions on ax2
ax2.set_xticks(ticks)
ax2.set_yticks(ticks)

# Plot data on ax and ax2
ax.plot(x, y)
ax2.plot(x, y)

# Set tick label font size on ax2
ax2.tick_params(axis='both', which='major', labelsize=8)

# Set tick positions on ax2
ax2.set_xticks(tick)

# Set tick label font size on ax
ax.tick_params(axis='both', which='major', labelsize=8)
```
